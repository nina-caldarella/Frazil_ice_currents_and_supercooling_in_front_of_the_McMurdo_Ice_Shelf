```python
# Import the required packages
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
from math import pi
import xarray as xr
# Uncomment the following line in case you are missing those packages
# !pip install windrose openpyxl
from windrose import WindroseAxes
```

```python
import os
os.chdir("../../..")
savedir = os.path.join(os.getcwd(),"DATA")
path = os.getcwd()
path = os.path.join(path,"DATA/SIOS21/adcp_data_and_analysis_lars_smedsrud/")
ds = xr.open_dataset(path+"Nortek_ADCP_currents_Oct20_to_Nov4_avgd_cal.nc")
```

```python
# DIR: The direction the ocean current is flowing toward. 0-360 degrees, 360 is due north, 0 means no measurable current.
# https://www.ndbc.noaa.gov/faq/measdes.shtml
DIR = xr.zeros_like(ds.v)
DIR = np.rad2deg(np.arctan2(ds.u, ds.v))
DIR[DIR < 0] += 360  # Wrap angles to 0–360 degrees
V_MAG = np.sqrt(ds.u.data**2+ds.v.data**2)
```

```python
# Compute mean vector components
mean_u = np.nanmean(ds.u.data)
mean_v = np.nanmean(ds.v.data)

# Compute average direction in degrees
mean_dir = np.rad2deg(np.arctan2(mean_u, mean_v))
if mean_dir < 0:
    mean_dir += 360

print(f"Average current direction: {mean_dir:.2f}°")
```

```python
ax = WindroseAxes.from_ax()
ax.bar(DIR, V_MAG, normed=True, opening=0.8, edgecolor='white', nsector=9*4)
ax.legend(title = "Current \n"+ r"magnitude (cm s$^{-1}$)", fontsize=14)
ax.set_title("Distribution of current direction and \n magnitude throughout time and for all bins")
plt.rcParams.update({'font.size': 14})
plt.savefig(os.path.join(savedir,"ADCP_direction.png"), dpi=300)
```
