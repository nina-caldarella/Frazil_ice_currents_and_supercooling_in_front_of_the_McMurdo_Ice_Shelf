```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# camera specs
sensor_diagonal = 7.2e-3 # m
f_stop = 3.8
focal_length = 26.8e-3 

# imaging geometry
object_distance = 1.45 # m

# calculate the circle of confusion
circle_of_confusion = sensor_diagonal / 1500 # Zeiss

# calculate the depth of field
DoF = (2*object_distance**2*f_stop*circle_of_confusion)/focal_length**2
print("DoF using Zeiss formula: ", DoF, " m")
# print("DoF in seawater: ", DoF*1.34, " m")
```
